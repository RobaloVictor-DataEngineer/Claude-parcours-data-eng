### Import + Series / DataFrame

```python
import pandas as pd        # convention : pandas sous l'alias "pd"

# Une Series = une seule colonne (1D) avec un index 0,1,2...
salaires = pd.Series([38000, 41000, 45000])
print(salaires)
```

```python
# Un DataFrame = un tableau ; chaque clé du dict = une colonne
df = pd.DataFrame({
    "nom": ["Alice", "Bob", "Chloé"],
    "salaire": [38000, 41000, 45000],
    "service": ["Data", "Data", "RH"],
})
df
```

### Lire un CSV (pour info) + inspecter

```python
# Sur un vrai fichier : df = pd.read_csv("data/employes.csv")
# Ici on inspecte le df créé au-dessus :
print(df.shape)     # (nb_lignes, nb_colonnes)
print(df.columns)   # noms de colonnes
print(df.dtypes)    # type de chaque colonne
df.describe()       # stats des colonnes numériques
```

### Sélectionner des colonnes

```python
df["salaire"]            # une colonne -> Series
```

```python
df[["nom", "salaire"]]   # plusieurs colonnes (liste de noms) -> DataFrame
```

### loc (étiquette) vs iloc (position)

```python
print(df.loc[0, "nom"])   # loc = par étiquette : ligne d'index 0, colonne "nom"
print(df.iloc[0, 1])      # iloc = par position : 1re ligne, 2e colonne
df.iloc[0:2]              # positions 0 et 1 (2 EXCLUE) ; avec loc, 0:2 inclurait 2
```

### Filtrage booléen

```python
df[df["salaire"] > 40000]                                  # une condition
```

```python
df[(df["salaire"] > 40000) & (df["service"] == "Data")]    # ET : () autour de chaque condition, & pour combiner
```

### Trier + colonne calculée

```python
df.sort_values("salaire", ascending=False)   # tri décroissant (renvoie une copie)
```

```python
df["salaire_annuel"] = df["salaire"] * 12    # nouvelle colonne sur toute la colonne, sans boucle
df
```
